# Linear plots with matplotlib

```python
import matplotlib.pyplot as plt

x = [7, 14, 21, 28, 35, 42, 49]
y = [8, 13, 21, 30, 31, 44, 50]
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


arr = np.random.randint(low=1, high=10, size=100)
plt.plot(arr, linewidth=3)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [7, 14, 21, 28, 35, 42, 49]
y = [8, 13, 21, 30, 31, 44, 50]

plt.plot(x, y, "or--", linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([7, 11, 24, 28, 35, 34, 41])
y = np.array([8, 20, 13, 30, 31, 48, 50])

plt.scatter(x, y, marker="*", color="r")
m, c = np.polyfit(x, y, 1)

print(m, c)

plt.plot(x, m*x + c, "g")
plt.show()
```

```python
# import
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

Name = np.array(["Ali", "Reza", "Hassan", "Sara", "Ayda"])
Score = np.array(["12.45", "19.65", "20.00", "16.56", "19.98"])

df = pd.DataFrame({
    "Name": Name,
    "Score": Score
})

plt.figure(figsize=(9, 5))

plt.plot(np.array(df["Name"]), np.array(df["Score"]), linewidth=2)

plt.xticks(rotation=60)
plt.xlabel("Name")
plt.ylabel("Score")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# creating the dataset
df = sns.load_dataset("iris")
df = df.groupby("sepal_length")["sepal_width"].sum().to_frame().reset_index()

plt.plot(np.array(df["sepal_length"]), np.array(df["sepal_width"]))


plt.title("This is title")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import numpy as np
from mpl_toolkits import mplot3d

# setting 3 axes for the graph
plt.axes(projection="3d")

z = np.linspace(0, 1, 100)

x = 4.5 * z
y = 0.8 * x + 2

plt.plot(x, y, z, "r", linewidth=2)
plt.title("Plot a 3D line")
plt.show()
```
